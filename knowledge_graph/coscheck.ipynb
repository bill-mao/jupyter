```python

import csv


rate_group = {}
# group com date    rate

with open("d:/desktop/check/rate_group.csv", encoding="utf8") as csvfile:
    spamreader = csv.reader(csvfile, quotechar='"')
    count = 0
    for cols in spamreader:
        if count == 0:
            count+=1
            continue
        else:
            group, com, date, rate = cols
            more = 6- len(str(com))
            com = '0'*more + com

            if group in rate_group.keys():
                rate_group[group] . append([ com+"-"+date , rate])
            else:
                rate_group[group] = [ [ com+"-"+date , rate]   ]

print(rate_group)

```
